```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_file='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv'
df=pd.read_csv(house_file)
df.head()
```

```python
df.dtypes
```

```python
df.describe(include='all')
```

```python
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
df.describe()
```

```python
print("total NaN values for the column bedrooms is  ", df['bedrooms'].isnull().sum())
print("total NaN values for the column bathrooms is ", df['bathrooms'].isnull().sum())
```

```python
#we need to fill the nan values first , in the beginning we will add the mean values of respective columns to corresponding nan values
```

```python
mean_bedrooms=df['bedrooms'].mean()
mean_bathrooms=df['bathrooms'].mean()
df['bedrooms'].replace(np.nan,mean_bedrooms, inplace=True)
df['bathrooms'].replace(np.nan,mean_bathrooms, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
new_cols =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
x = df[new_cols]
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
print('test samples : ',x_test.shape[0])
```

```python
print('train samples : ',x_train.shape[0])
```

```python
from sklearn.linear_model import Ridge
```

```python
rm = Ridge(alpha = 0.1)
rm.fit(x_train, y_train)
rm.score(x_test, y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
```

```python
pr = PolynomialFeatures(degree=2)
pr
```

```python
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
poly = Ridge(alpha=0.1)
poly.fit(x_train_pr, y_train)
poly.score(x_test_pr, y_test)
```
